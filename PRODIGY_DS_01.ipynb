```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data (ages and genders of individuals)
np.random.seed(0)  # for reproducibility
ages = np.random.randint(20, 70, size=40)
genders = np.random.choice(['Male', 'Female'], size=40)

# Separate ages by gender
male_ages = ages[genders == 'Male']
female_ages = ages[genders == 'Female']

# Define bin edges
bin_edges = np.arange(20, 71, 5)  # Bin edges from 20 to 70 with a bin width of 5

# Create a figure for male histogram
plt.figure()
plt.hist(male_ages, bins=bin_edges, color='ORANGE', edgecolor='black', alpha=0.5, label='Male', align='mid')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages for Males')
plt.legend()

# Create a figure for female histogram
plt.figure()
plt.hist(female_ages, bins=bin_edges, color='PURPLE', edgecolor='black', alpha=0.5, label='Female', align='mid')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages for Females')
plt.legend()

# Show plots
plt.show()
```
